### Werte angleichen 
Teilweise haben Nullwerte verschiedene Bezeichnungen oder auch andere Werte unterscheiden sich durch Groß- und Kleinschreibung. Diese werden nun angepasst.  
  
Zuerst werden die Umlaute und anderen Sonderzeichen entfernt.


```python
# Großbuchstaben angleichen
    # nicht notwendig, alles ist bereits in Großbuchstaben
# Funktion für Großbuchstaben in Spaltenbezeichnung
def convert_columns_to_uppercase(df):
    df.columns = [col.upper() for col in df.columns]
    return df

# Sonderzeichen entfernen
from unidecode import unidecode
# Funktion zum Entfernen der Sonderzeichen
def remove_accents(df):
    return df.apply(lambda col: col.map(lambda x: unidecode(x) if isinstance(x, str) else x))
# Funktion zum Entfernen der Sonderzeichen in den Spaltennamen
def remove_accents_from_columns(df):
    df.columns = [unidecode(col) for col in df.columns]
    return df

# Slash und Backslash in Daten durch Unterstrich ersetzen
def replace_slashes(df):
    def replace_func(x):
        if isinstance(x, str):
            return x.replace('/', '_').replace('\\', '_')
        return x
    
    # Ersetzen von '/' und '\' durch '_'
    df = df.apply(lambda col: col.map(replace_func))
    return df
```

```python
# Anwenden der Funktionen auf den gesamten DataFrame
data_all_2020 = remove_accents(data_all_2020)
data_all_2020 = remove_accents_from_columns(data_all_2020)
data_all_2020 = convert_columns_to_uppercase(data_all_2020)
data_all_2020 = replace_slashes(data_all_2020)
```

**Ergebnis**: keine Spalten mit nur einem Wert  
Deswegen 3. Delete Columns That Contain a Single Value nicht notwendig.
  
**Dafür**: 
* Viele Spalten mit unterschiedlichen Ausdrücken für Nullwerte
* Dopplung von Werten durch unterschiedliche Ausdrücke

**Entscheidung**:
In vielen Kategorien gibt es den Wert 'Não sabe/soube informar' ("weiß es nicht"). Obwohl dessen Informationsgehalt höher ist als NaN, da in erster Linie ausgesagt wird, dass die anrufende Person darüber keine Aussage treffen kann, wird dieser Wert entfernt. Der Informationsgehalt ist zu niedrig, weshalb im ersten Anlauf ohne diesen Wert modelliert wird.  
Ebsenso wird 'Atendimento interrompido' ("Anruf abgebrochen") als NaN behandelt.
Ebenso 'Nao se aplica'

```python
# Funktion ersetzt alle Vorkommen eines bestimmten Werts im DataFrame durch NaN
def replace_value_with_nan(df, value):
    return df.replace(value, np.nan)

# Funktion zur Eliminierung des Prefixes 'BR | ' oder '520870 | ' bei Ortsangaben
def remove_prefix(df, delimiter=' | '):
    def strip_prefix(x):
        if isinstance(x, str) and delimiter in x:
            return x.split(delimiter, 1)[1].strip()
        return x
    return df.apply(lambda col: col.map(strip_prefix) if col.dtype == 'O' else col)

# Funktion zum Ersetzen eines bestimmten Werts durch einen anderen Wert
def replace_value(df, old_value, new_value):
    return df.replace(old_value, new_value)

# Funktion zum Aufteilen der Werte in drei Spalten
def split_column(df, column_name, delimiter='>', new_columns=None):
    if new_columns is None:
        new_columns = [f"{column_name}_{i}" for i in range(1, 4)]
    
    # Split the column into multiple columns
    split_data = df[column_name].str.split(delimiter, expand=True)

    # Ensure there are at least three columns
    if split_data.shape[1] < 3:
        for i in range(split_data.shape[1], 3):
            split_data[i] = None

    # Combine any extra columns into the third column
    split_data[2] = split_data.iloc[:, 2:].apply(lambda x: ', '.join(x.dropna()), axis=1)
    
    # Only keep the first three columns
    split_data = split_data.iloc[:, :3]
    split_data.columns = new_columns

    # Join the new columns back to the original dataframe
    df = df.drop(columns=[column_name])
    df[new_columns] = split_data

    return df
```
